```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = pd.read_csv('Country-data.csv')
```

```python
# Display the first few rows of the dataframe
df.head()
```

```python
# Check for duplicated rows
df.duplicated().sum()
```

```python
# Check for missing values in each column
df.isnull().sum()
```

```python
# Get the shape of the dataframe
df.shape
```

```python
# Make a copy of the dataframe
df_cpy = df.copy()
```

```python
# Drop the 'country' column from the dataframe
df = df.drop(columns='country')
df.head()
```

```python
# Explore top countries with highest child mortality
df_cpy.sort_values('child_mort', ascending=False).head(10)
```

```python
# Explore top countries with lowest GDP per capita
df_cpy.sort_values('gdpp', ascending=True).head(10)
```

```python
# Explore top countries with highest inflation
df_cpy.sort_values('inflation', ascending=False).head(10)
```

```python
# Explore top countries with lowest health expenditure
df_cpy.sort_values('health', ascending=True).head(10)
```

```python
# Adjust health, exports, and imports columns based on GDP
df_cpy.health = ((df_cpy.health * df_cpy.gdpp) / 100)
df_cpy.exports = ((df_cpy.exports * df_cpy.gdpp) / 100)
df_cpy.imports = ((df_cpy.imports * df_cpy.gdpp) / 100)
```

```python
# Explore top countries with highest and lowest imports and exports
df_cpy.sort_values('imports', ascending=False).head(10)
df_cpy.sort_values('exports', ascending=True).head(10)
```

```python
# Explore top countries with lowest life expectancy and highest fertility rate
df_cpy.sort_values('life_expec', ascending=True).head(10)
df_cpy.sort_values('total_fer', ascending=False).head(10)
```

```python
# Drop the 'country' column from the dataframe
df = df_cpy.drop(columns='country')
```

```python
# Visualize relationships between variables using pairplot
sns.pairplot(data=df)
```

```python
# Calculate correlation matrix
corr_matrix = df.corr()
```

```python
# Visualize correlation matrix using heatmap
sns.heatmap(data=corr_matrix, cmap='RdBu_r', annot=True)
```
